```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#uniform distribution
req_size_arrays = [10, 100, 500, 1000, 5000, 10000, 50000]
req_arrays_uniform = {i:np.random.uniform(0,1000000,i) for i in req_size_arrays}
req_arrays_uniform_mean = {str(i):np.mean(req_arrays_uniform[i]) for i in req_arrays_uniform}
plt.plot(req_arrays_uniform_mean.keys(),req_arrays_uniform_mean.values())
plt.show()
```

```python
#exponential distribution
req_size_arrays = [10, 100, 500, 1000, 5000, 10000, 50000]
req_arrays_exponential = {i:np.random.exponential(scale=1,size=i) for i in req_size_arrays}
req_arrays_exponential_mean = {str(i):np.mean(req_arrays_exponential[i]) for i in req_arrays_exponential}
plt.plot(req_arrays_exponential_mean.keys(),req_arrays_exponential_mean.values())
plt.show()
```

```python
#bernoulli distribution
req_arrays_bernoulli = {i:np.random.binomial(1,0.5,i) for i in req_size_arrays}
req_arrays_bernoulli_mean = {str(i):np.mean(req_arrays_bernoulli[i]) for i in req_arrays_bernoulli}
plt.plot(req_arrays_bernoulli_mean.keys(),req_arrays_bernoulli_mean.values())
plt.show()
```

```python
# in all the three graphs given above you can see that the means are stabilizing towards the expected mean value hence
# verifying the weak law of large numbers
```
