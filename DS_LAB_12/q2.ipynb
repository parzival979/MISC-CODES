```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Exponential Distribution
req_arrays_exponential = {i:np.random.exponential(1,size=1000) for i in range(2)}
req_arrays_exponential_mean = {str(i):np.mean(req_arrays_exponential[i]) for i in req_arrays_exponential}
sns.histplot(req_arrays_exponential_mean.values(),kde=True)
```
