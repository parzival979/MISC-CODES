## USING SKLEARN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Loading the dataset
df = pd.read_csv('sonar.all-data.csv', header=None)
print(df.head())
```

```python
# removing the labels
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
print(X.head())
print(y.head())
```

```python
# splitting the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# KNN with K=1
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
```

```python
# KNN with K=3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
```

```python
# KNN with K=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
```

```python
# KNN with K=7
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
```

```python
# creating a matrix to store accuracies for 6 subsets
acc = np.zeros((6, 4))
```

```python
# KNN After breaking into 6 subsets
for i in range(1, 7):
    for j in range(1, 8, 2):
        knn = KNeighborsClassifier(n_neighbors=j)
        knn.fit(X_train.iloc[:, (i-1)*10:i*10], y_train)
        y_pred = knn.predict(X_test.iloc[:, (i-1)*10:i*10])
        print('Accuracy: ', accuracy_score(y_test, y_pred))
        print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
        # storing the accuracy
        acc[i-1, int(j/2)] = accuracy_score(y_test, y_pred)
# printing the accuracies
print(acc)
```

```python
# plotting the accuracies
plt.plot([1, 3, 5, 7], acc[0], label='Subset 1')
plt.plot([1, 3, 5, 7], acc[1], label='Subset 2')
plt.plot([1, 3, 5, 7], acc[2], label='Subset 3')
plt.plot([1, 3, 5, 7], acc[3], label='Subset 4')
plt.plot([1, 3, 5, 7], acc[4], label='Subset 5')
plt.plot([1, 3, 5, 7], acc[5], label='Subset 6')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# creating an array to store the accuracies of majority voting
acc = np.zeros(4)
```

```python
# function for KNN using sklearn
def KNN(X_train, X_test, y_train, y_test, k):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    return y_pred
```

```python
# KNN After majority voting
# performing majority voting with 6 subsets for K = 1
y_pred = []
for i in range(len(X_test)):
    y_pred.append(pd.Series([KNN(X_train.iloc[:, (j-1)*10:j*10], X_test.iloc[i:i+1, (j-1)*10:j*10], y_train, y_test, 1)[0] for j in range(1, 7)]).value_counts().index[0])
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
# storing the accuracy
acc[0] = accuracy_score(y_test, y_pred)
```
