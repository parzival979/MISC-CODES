```python
import PIL
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the image from img.png
img = Image.open('img.png')
```

```python
img
```

```python

# Convert the image to a numpy array
img_array = np.array(img)
print(img_array)
```

```python
# extract the red channel
red = img_array[:,:,0]
```

```python
# extract a 32x32 patch from the red channel
patch = red[0:32,0:32]
print(patch)
```

```python
# plot the patch
plt.imshow(patch, cmap='gray')
plt.show()
```

```python
# extract the maximum pixel values along each row and column and print them
max_row = np.max(patch, axis=1)
max_col = np.max(patch, axis=0)
print(max_row)
print(max_col)
```

```python
# generate a new patch whose pixel values are the cosine of the original patch and display it
patch_cos = np.cos(patch)
plt.imshow(patch_cos, cmap='gray')
plt.show()
```
