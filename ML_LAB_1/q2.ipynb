```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 300 3-dimensional vectors that come from a normal distribution with mean vector as [1 2 1]t and 3x3 covariance matrix as [5 0.8 -0.3; 0.8 3 0.6; -0.3 0.6 4]
mean = [1, 2, 1]
cov = [[5, 0.8, -0.3], [0.8, 3, 0.6], [-0.3, 0.6, 4]]
data = np.random.multivariate_normal(mean, cov, 300)
```

```python
# make scatter plot of x1 vs x2
x1 = data[:, 0]
x2 = data[:, 1]
x3 = data[:, 2]
plt.scatter(x1, x2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# make scatter plot of x1 vs x3
plt.scatter(x1, x3)
plt.xlabel('x1')
plt.ylabel('x3')
plt.show()
```

```python
# make scatter plot of x2 vs x3
plt.scatter(x2, x3)
plt.xlabel('x2')
plt.ylabel('x3')
plt.show()
```

```python
# # Compute the mean vector and covariance matrix of the data
mean = np.mean(data, axis=0)
cov = np.cov(data, rowvar=False)
print('mean vector: ', mean)
print('covariance matrix: ', cov)
```
