```python
import PIL
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the dataframe
df = pd.read_csv('Auto.csv')
```

```python
# use groupby to calculate the mean of mpg for each cylinder
df.groupby('cylinders').mean()['mpg']
```

```python
# generate scatter plot of mpg vs displacement
plt.scatter(df['displacement'], df['mpg'])
plt.xlabel('displacement')
plt.ylabel('mpg')
plt.show()
```
