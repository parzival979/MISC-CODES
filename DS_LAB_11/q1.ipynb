```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
req_df=pd.read_csv('landslide_data3_miss.csv')
req_dict_missing = {}
for i in req_df.columns:
    req_dict_missing[i]=req_df[i].isnull().sum()
print(req_dict_missing)
```

```python
plt.bar(req_dict_missing.keys(),req_dict_missing.values())
plt.xticks(rotation=45)
plt.show()
```

```python
req_df=req_df.dropna(axis=0,subset=['stationid'])
print(req_df)
init_len = len(req_df)
```

```python
print("Number of Tuples Deleted From StationID: "+str(req_dict_missing['stationid']))
```

```python
req_df=req_df.dropna(axis=0,thresh=7)
final_len = len(req_df)
print(req_df)
```

```python
print("The Number Of rows deleted with nulls 3 and more are: "+str(init_len-final_len))
```

```python
for i in req_df.columns:
    req_dict_missing[i]=req_df[i].isnull().sum()
print(req_dict_missing)
plt.bar(req_dict_missing.keys(),req_dict_missing.values())
plt.xticks(rotation=45)
plt.show()
```

```python
print("The Total Number Of Missing Values is: "+str(sum(req_dict_missing.values())))
```
