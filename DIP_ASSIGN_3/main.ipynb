```python
# importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# importing the image
img = cv2.imread('Lena.png')
```

```python
print(img)
```

```python
# converting the image to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
print(img)
```

```python
# displaying the image
plt.imshow(img)
plt.show()
```

```python
# converting the image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
print(img_gray)
```

```python
# displaying the grayscale image
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# function to return the resized image using cv2.resize() and dimensions
def resize_image(img, dim):
    return cv2.resize(img, dim)
```

```python
# creating 128,64,32 and 16 dimensional images in different variables
img_128 = resize_image(img_gray, (128,128))
img_64 = resize_image(img_gray, (64,64))
img_32 = resize_image(img_gray, (32,32))
img_16 = resize_image(img_gray, (16,16))
```

```python
# displaying all the images in a single plot
fig, ax = plt.subplots(2,2, figsize=(10,10))
ax[0,0].imshow(img_128,cmap='gray')
ax[0,0].set_title('128x128')
ax[0,1].imshow(img_64,cmap='gray')
ax[0,1].set_title('64x64')
ax[1,0].imshow(img_32,cmap='gray')
ax[1,0].set_title('32x32')
ax[1,1].imshow(img_16,cmap='gray')
ax[1,1].set_title('16x16')
plt.show()
```

#### You can clearly see that as the resolution of the image decreases the detail decreases and the image becomes distorted.

```python
# function to return the downsampled image given the image and the intensity range of 2 numbers
def downsample_image(img, intensity_range):
    return img//intensity_range
```

```python
# creating 8 different intensity range images
img_256 = downsample_image(img_gray, 1)
print(img_256.shape)
img_128 = downsample_image(img_gray, 2)
print(img_128)
img_64 = downsample_image(img_gray, 4)
print(img_64)
img_32 = downsample_image(img_gray, 8)
img_16 = downsample_image(img_gray, 16)
img_8 = downsample_image(img_gray, 32)
img_4 = downsample_image(img_gray, 64)
img_2 = downsample_image(img_gray, 128)
```

```python
# displaying all the images in a single plot
fig, ax = plt.subplots(2,4, figsize=(20,10))
ax[0,0].imshow(img_256,cmap='gray')
ax[0,0].set_title('256')
ax[0,1].imshow(img_128,cmap='gray')
ax[0,1].set_title('128')
ax[0,2].imshow(img_64,cmap='gray')
ax[0,2].set_title('64')
ax[0,3].imshow(img_32,cmap='gray')
ax[0,3].set_title('32')
ax[1,0].imshow(img_16,cmap='gray')
ax[1,0].set_title('16')
ax[1,1].imshow(img_8,cmap='gray')
ax[1,1].set_title('8')
ax[1,2].imshow(img_4,cmap='gray')
ax[1,2].set_title('4')
ax[1,3].imshow(img_2,cmap='gray')
ax[1,3].set_title('2')
```

#### You can clearly see that as the intensity range decreases the image becomes more and more less detailed as the amount of information is reduced.
