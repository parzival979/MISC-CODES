```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the data with seperator as space
data = pd.read_csv('data.csv', sep = ' ', header = None)
print(data)
```

```python
# performing PCA on the data without sklearn
X = np.array(data)
Xmean = np.mean(X, axis = 0)
print(Xmean)
```

```python
# finding covariance matrix
cov = np.cov(X.T)
print(cov)
```

```python
# finding eigen values and eigen vectors
eig_val, eig_vec = np.linalg.eig(cov)
print("The eigen values are: ")
print(eig_val)
print("The eigen vectors are: ")
print(eig_vec)
```

```python
# sorting the eigen values and eigen vectors
idx = eig_val.argsort()[::-1]
eig_val = eig_val[idx]
eig_vec = eig_vec[:,idx]
print("The sorted eigen values are: ")
print(eig_val)
print("The sorted eigen vectors are: ")
print(eig_vec)
```

```python
# finding the projection matrix for 2D
P = eig_vec[:,0:2]
print("The projection matrix is: ")
print(P)
```

```python
# finding the projected data
Xproj = np.dot(X-Xmean,P)
print("The projected data is: ")
print(Xproj)
```

```python
# finding the projection matrix for 3D
P = eig_vec[:,0:3]
print("The projection matrix is: ")
print(P)
```

```python
# finding the projected data
Xproj = np.dot(X-Xmean,P)
print("The projected data is: ")
print(Xproj)
```

```python
# plotting the data in 2D for X1 and X2
plt.scatter(Xproj[:,0], Xproj[:,1])
# adding labels
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('2D plot for X1 and X2')
plt.show()
```

```python
# plotting the data in 2D for X1 and X3
plt.scatter(Xproj[:,0], Xproj[:,2])
# adding labels
plt.xlabel('X1')
plt.ylabel('X3')
plt.title('2D plot for X1 and X3')
plt.show()
```

```python
# plotting the data in 2D for X2 and X3
plt.scatter(Xproj[:,1], Xproj[:,2])
# adding labels
plt.xlabel('X2')
plt.ylabel('X3')
plt.title('2D plot for X2 and X3')
plt.show()
```

```python
# plotting the data in 3D
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Xproj[:,0], Xproj[:,1], Xproj[:,2])
# adding labels
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
plt.title('3D plot for X1, X2 and X3')
plt.show()
```

```python
# performing inverse PCA on the data without sklearn
Xinv = np.dot(Xproj, P.T)+Xmean
print("The inverse projected data is: ")
print(Xinv)
```

```python
# finding mean squared error
mse = np.mean((X-Xinv)**2)
print("The mean squared error is: ")
print(mse)
```
