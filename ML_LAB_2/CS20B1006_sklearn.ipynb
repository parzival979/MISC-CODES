```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Loading the data with seperator as space
data = pd.read_csv('data.csv', sep = ' ', header = None)
print(data)
```

```python
# printing the eigen vectors for the data
pca = PCA(n_components = 4)
pca.fit(data)
print(pca.components_)

# printing the explained variance(eigen values) for the data
print(pca.explained_variance_)
```

```python
# performing PCA on the data for 2 dimensions
pca = PCA(n_components = 2)
pca.fit(data)
transformed_data_2 = pca.transform(data)
print(transformed_data_2)

# printing the eigen vectors
print("Eigen Vectors are: ")
print(pca.components_)
```

```python
# performing PCA on the data for 3 dimensions
pca = PCA(n_components = 3)
pca.fit(data)
transformed_data_3 = pca.transform(data)
print(transformed_data_3)

# printing the eigen vectors
print("Eigen Vectors are: ")
print(pca.components_)
```

```python
# plotting the features in the scatter plot from the transformed data x1 and x2
plt.scatter(transformed_data_2[:, 0], transformed_data_2[:, 1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('Scatter plot of the data in 2 dimensions')
plt.show()
```

```python
# plotting the features in the scatter plot from the transformed data x1 and x3
plt.scatter(transformed_data_3[:, 0], transformed_data_3[:, 2])
plt.xlabel('First Principal Component')
plt.ylabel('Third Principal Component')
plt.title('Scatter plot of the data in 2 dimensions')
plt.show()
```

```python
# plotting the features in the scatter plot from the transformed data x2 and x3
plt.scatter(transformed_data_3[:, 1], transformed_data_3[:, 2])
plt.xlabel('Second Principal Component')
plt.ylabel('Third Principal Component')
plt.title('Scatter plot of the data in 2 dimensions')
plt.show()
```

```python
# plotting the features in the scatter plot from the transformed data x1, x2 and x3
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(transformed_data_3[:, 0], transformed_data_3[:, 1], transformed_data_3[:, 2])
ax.set_xlabel('First Principal Component')
ax.set_ylabel('Second Principal Component')
ax.set_zlabel('Third Principal Component')
plt.title('Scatter plot of the data in 3 dimensions')
plt.show()
```

```python
# performing inverse transformation to get original data
inverse_transformed_data = pca.inverse_transform(transformed_data_3)
print(inverse_transformed_data)
```

```python
# calculating the mean square error
mean_square_error = np.mean((np.array(data) - inverse_transformed_data)**2)
print("The mean square error is: ")
print(mean_square_error)
```
