Create a concentric square image, with inner square pixel intensity values as 125 and outer
square as 0 everywhere. Find the projection along the row, column, diagonal with 45 degree
and 135 degree. Reconstruct the image by applying back projection algorithm using the
following ways:
1. the row projection only
2. the row and column projections
3. the row and column and diagonal 45-degree projections
4. the row and column and diagonal 45- and 135-degree projections

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# creating the concentric square image
img = np.zeros((100,100))
img[25:75,25:75] = 125
plt.imshow(img, cmap='gray')
plt.title('Concentric Square Image')
plt.show()
```

```python
# finding the projection along the row
row_proj = np.sum(img, axis=1)
plt.plot(row_proj)
plt.title('Row Projection')
plt.show()
```

```python
# finding the projection along the column
col_proj = np.sum(img, axis=0)
plt.plot(col_proj)
plt.title('Column Projection')
plt.show()
```

```python
# finding the projection along the diagonal with 45 degree
diag_proj1 = np.zeros(100)
for i in range(100):
    diag_proj1[i] = np.sum(img.diagonal(i))
plt.plot(diag_proj1)
plt.title('Diagonal Projection with 45 degree')
plt.show()
```

```python
# finding the projection along the diagonal with 135 degree
diag_proj2 = np.zeros(100)
for i in range(100):
    diag_proj2[i] = np.sum(np.fliplr(img).diagonal(i))
plt.plot(diag_proj2)
plt.title('Diagonal Projection with 135 degree')
plt.show()
```

```python
# reconstructing the image by applying back projection algorithm using the row projection only
img_back = np.zeros((100,100))
for i in range(100):
    img_back[i,:] = row_proj[i]
plt.imshow(img_back, cmap='gray')
plt.title('Reconstructed Image using Row Projection')
plt.show()
```

```python
# reconstructing the image by applying back projection algorithm using the row and column projections
img_back = np.zeros((100,100))
for i in range(100):
    for j in range(100):
        img_back[i,j] = (row_proj[i] + col_proj[j])//2
plt.imshow(img_back, cmap='gray')
plt.title('Reconstructed Image using Row and Column Projections')
plt.show()
```

```python
# reconstructing the image by applying back projection algorithm using the row and column and diagonal 45-degree projections
img_back = np.zeros((100,100))
center = 100//2

for i in range(100):
    for j in range(100):
        if i==j:
            img_back[i,j] = diag_proj1[center]
        else:
            img_back[i,j] = (row_proj[i] + col_proj[j] + diag_proj1[center])//3
plt.imshow(img_back, cmap='gray')
plt.title('Reconstructed Image using Row, Column and Diagonal 45-degree Projections')
plt.show()
```

```python
# reconstructing the image by applying back projection algorithm using the row and column and diagonal 45- and 135-degree projections
img_back = np.zeros((100,100))
for i in range(100):
    for j in range(100):
        if i==j:
            img_back[i,j] = (diag_proj1[center] + diag_proj2[center])//2
        else:
            img_back[i,j] = (row_proj[i] + col_proj[j] + diag_proj1[center] + diag_proj2[center])//4
plt.imshow(img_back, cmap='gray')
plt.title('Reconstructed Image using Row, Column and Diagonal 45- and 135-degree Projections')
plt.show()
```
